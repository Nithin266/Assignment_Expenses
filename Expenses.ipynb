# Feature Engineering

### In feature engineering we use more techniques to remove null value or fill NaN values
### And more features like label encoding replacing the values

```python
data.head()
```

```python
data1=data.copy() ## Take the copy of data
```

```python
data1.info()
```

```python
data1.columns
```

```python
data1.head()
```

```python
#### Using dummies feature to replace the word to numbers to implement the machine learning model

data1 = pd.get_dummies(data1,columns=["sex", "smoker", "region"],drop_first=True)
```

```python
data1.head()
```

```python
data1.info()
```

# Feature Selection

### In the feature selection part we can appy more machine learning model to get good and accurate model

```python
data1.head()
```

```python
## setting dependent and independent features
X=data1.drop(["charges"],axis=1)
y= data1["charges"]
```

```python
X.head()
```

```python
y.head()
```

```python
#### Splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_1=LinearRegression()
```

```python
model_1.fit(X_train,y_train)
```

```python
model_1.score(X_test,y_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=3,min_samples=4)
```

```python
model_2=dbscan.fit(X_train)
label=model_2.labels_
```

```python
from sklearn import metrics

#identifying the points which makes up our core points
sample_cores=np.zeros_like(label,dtype=bool)

sample_cores[dbscan.core_sample_indices_]=True

#Calculating the number of clusters

n_clusters=len(set(label))- (1 if -1 in label else 0)



print(metrics.silhouette_score(X_train,label))
```
